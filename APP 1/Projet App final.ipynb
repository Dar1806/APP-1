```python
#Librairie
import pandas as pd
import sys
import customtkinter
from customtkinter import * 
from tkinter import *
from tkinter import ttk
import numpy as np
from PIL import Image
from tkinter import filedialog as fd
from tkinter.messagebox import showinfo


#Paramètres variables
pd.set_option('display.max_columns',None)
df = []
lst = []
val = []


#Paramètres APP
app = customtkinter.CTk()
app.title("360 Product-Application")
app.geometry("1150x1100")
set_appearance_mode("light")


#Fonction recherchant le document souhaité
def select_files():
    global df,lst,val
    filetypes = (('Excel files', '*.xlsx'),('All files', '*.*'))
    filename = fd.askopenfilename(title='Open files',initialdir='/',filetypes=filetypes)
    if filename :
        try :
            df = pd.read_excel(filename)
            showinfo(title='Selected Files',message=filename)
            lst = df["Fund name"]
            lst = list(set(lst))
            val = df.columns
        except Exception as e :
            showinfo(title='Selected Files',message=filename)
            df = pd.read_excel(filename)
            showinfo(title='Selected Files',message=filename)
            print(df)
            lst = df["Fund name"]
            lst = list(set(lst))
            val = df.columns


#Paramètre les fonctions en global 
result_share = None
global fonds_trouves 
fonds_trouves = None
value = None


#Image de fond 



fond_image = customtkinter.CTkImage(Image.open("Coupole_03.jpg"),size=(1400,1100)) 
image_button = customtkinter.CTkLabel(master=app, text="",image=fond_image)
image_button.pack()
image_button.place(x=0, y=0)


sg_image = customtkinter.CTkImage(Image.open("logo-societe-generale.jpg"),size=(250,80)) 
image_sg = customtkinter.CTkLabel(master=app, text="",image=sg_image)
image_sg.pack()
image_sg.place(x=865, y=990)


fond_carre = customtkinter.CTkImage(Image.open("blanc.png"),size=(600,1050)) 
image_carre = customtkinter.CTkLabel(master=app, text="",image=fond_carre)
image_carre.pack()
image_carre.place(x=20, y=20)


selected = customtkinter.StringVar(value="")
#Fonction qui prends la valeur du widget combo
def get_combo(choice):
    global fonds_trouves
    xx=combo.get()
    liste_share= df[df['Fund name']==xx]["Share name"].tolist()

#Créer pour chaque parts une checkbox avec le nom de cette part dans frame
    for item in liste_share :
        fonds_trouves = customtkinter.CTkRadioButton(frame, text=item,height=10, width=80,
                                              font=("helvetica",12),variable=selected,value=item,
                                              command=lambda item=item : open(item),
                                              border_color="#383838",hover_color="#4E4E4E", fg_color="#4A8E3C",corner_radius=30,
                                              radiobutton_height=30,radiobutton_width=30,border_width_checked=7,border_width_unchecked=5)
        fonds_trouves.pack(pady=10,anchor=W)
        global fonds

def open(resultat_choix_fonds):
    global fonds
    fonds = resultat_choix_fonds


#Prends la valeur de la combobox des parts  
def get_share(choice):
    global info_trouves
    xx=radio_info.get()
    print(xx)
    info_trouves = customtkinter.CTkRadioButton(frame2, text=xx,height=10, width=80,
                                              font=("helvetica",12),variable=selected,value=xx,
                                              command=lambda item=xx : openn(item),
                                              border_color="#383838",hover_color="#4E4E4E", fg_color="#4A8E3C",corner_radius=30,
                                              radiobutton_height=30,radiobutton_width=30,border_width_checked=7,border_width_unchecked=5)
    info_trouves.pack(pady=10,anchor=W)
    global info
def openn(resultat_choix_info):
    global info
    info = resultat_choix_info
    print(info)
#Fonction recherche dans le combobox des Fonds 
def search(event):
    value = event.widget.get()
    if value == '':
        combo.configure (values = lst)
    
    else:
        data = []
        for item in lst : 
            if value.lower() in item.lower():
                data.append(item)
        combo.configure(values = data)


#Fonction recherche dans le combobox des Parts 
def radio(valo):
    global val
    value_rad = valo.widget.get()
    if value_rad == '':
        radio_info.configure (values = val)
    
    else:
        data = []
        for item in val : 
            if value_rad.lower() in item.lower(): 
                data.append(item)
        radio_info.configure(values = data)


#Fonction affichant la valeur cherchée 
def button_final():
    global fonds
    global value
    global info
    value =(df[df['Share name']==fonds][info].values[0])
    if pd.isna(value):
        info_finale.configure(text="No value in this cell  ")
    else :
        info_finale.configure(text=value)
        info_finale.configure(value=value)


#Fonction copiant la valeur finale dans le presse papier
def copy_to_clipboard():
    app.clipboard_clear()
    valuee = df[df['Share name']==fonds][info].values[0]
    app.clipboard_append(valuee)
    buttonff.configure(fg_color="#59A03D",hover_color="#67BB46",text="✅")
    buttonff.after(2000,  lambda : buttonff.configure(fg_color="#3C3C3C",hover_color="#4E4E4E",text="Copy"))


#INTERFACE TKINTER



#Widget Bouton permettant de choisir le fichier excel
open_button = customtkinter.CTkButton(app,text='Open Files',command=select_files,width=200,height=40,fg_color="#3C3C3C",
                                        corner_radius=6,hover_color="#4E4E4E", text_color="#FFFFFF",)
open_button.pack(expand=True)
open_button.place(x=50,y=90)


#Widget Frame permettant de stocker les checkboxs Shares dans une frame
cherche_xcl = customtkinter.CTkLabel(app, text="Chose your file :", font=("helvetica",16))
cherche_xcl.place(x=50,y=50)


#Widget Frame permettant de stocker les checkboxs Shares dans une frame
frame = customtkinter.CTkScrollableFrame(app, height=0, width=510, border_color="#484646",
                                        border_width=2, fg_color="#C4C3C3", corner_radius=10,
                                        label_text=("Share(s) Found "), label_fg_color="#3C3C3C", scrollbar_button_color="#3C3C3C", scrollbar_button_hover_color="#4E4E4E", label_text_color="#FFFFFF")
frame.place(x=50, y=240)


#Widget Label avec comme valeur "Search your found :"
cherche_fond = customtkinter.CTkLabel(app, text="Search your Fund :", font=("helvetica",16))
cherche_fond.place( x=50, y=150)


#Widget Combobox permettant de chercher tout les fonds de la liste du tableau excel 
combo= customtkinter.CTkComboBox(app,values=list(lst), width=400,height=40,
                                 button_color="#3C3C3C",button_hover_color="#4E4E4E",
                                 command=get_combo)
combo.set("")
combo.pack(pady=10)
combo.bind("<KeyRelease>", lambda event : get_combo(combo.get())+search(event))
combo.place(x=50,y=190)


#Widget Bouton permettant de supprimer tous les logs des Shares affichées dans frame
boutton_clear = customtkinter.CTkButton(app,text = 'Clear',width=110,height=40,fg_color="#3C3C3C",
                                        corner_radius=6,hover_color="#4E4E4E", text_color="#FFFFFF",
                                        command=lambda: [widget.destroy() for widget in frame.winfo_children()])
boutton_clear.pack(pady=10)
boutton_clear.place(x=480,y=190)


#Widget Bouton permettant de supprimer tous les logs des Shares affichées dans frame2
boutton_clear = customtkinter.CTkButton(app,text = 'Clear',width=110,height=40,fg_color="#3C3C3C",
                                        corner_radius=6,hover_color="#4E4E4E", text_color="#FFFFFF",
                                        command=lambda: [widget.destroy() for widget in frame2.winfo_children()])
boutton_clear.pack(pady=10)
boutton_clear.place(x=480,y=580)


#Widget Label avec comme valeur "What do you wanna know ? :"
cherche_info = customtkinter.CTkLabel(app, text="What do you want to know ? :", font=("helvetica",16))
cherche_info.place(x=50,y=535)


#Widget Combobox permettant de chercher les informations des colonnes excels 
radio_info = customtkinter.CTkComboBox(app,values=val, width=400,height=40,button_color="#3C3C3C",
                                       button_hover_color="#4E4E4E",command=get_share)
radio_info.set("")
radio_info.pack()
radio_info.bind("<KeyRelease>", radio)
radio_info.place(x=50, y=580)

buttonf = customtkinter.CTkButton(app,text = 'Submit',command=button_final,width=230,height=40,fg_color="#3C3C3C", text_color="#FFFFFF",
                                  corner_radius=6,hover_color="#4E4E4E",font=("helvetica",16))
buttonf.pack()
buttonf.place(x=190,y=930)

#Widget Label avec comme valeur 'Result :""
result = customtkinter.CTkLabel(app, text="Result :", font=("helvetica",16))
result.place(x=50,y=1000)


#Widget Label affichant la valeur souhaité 
info_finale = customtkinter.CTkLabel(app, text="", font=("helvetica",16),)
info_finale.place(x=110,y=1000)   

buttonff = customtkinter.CTkButton(app,text = 'Copy result',command=copy_to_clipboard,width=110,height=40,fg_color="#3C3C3C", text_color="#FFFFFF",
                                  corner_radius=6,hover_color="#4E4E4E")
buttonff.pack()
buttonff.place(x=480,y=990)


frame2 = customtkinter.CTkScrollableFrame(app, height=2, width=510, border_color="#484646",
                                        border_width=2, fg_color="#C4C3C3", corner_radius=10,
                                        label_text=("Log Info"), label_fg_color="#3C3C3C", scrollbar_button_color="#3C3C3C", scrollbar_button_hover_color="#4E4E4E", label_text_color="#FFFFFF")
frame2.place(x=50, y=640)

selected_items = []
checkboxes = {}
app.mainloop()
```
